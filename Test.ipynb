```python
import sys
import time
import random
import math
import os
import platform
import traceback
import numpy as np
import multiprocessing as mp
```

```python
global x
x = 10

def task1(l):
    l1 = l
    l2 = l
    global x
    l1.acquire()
    print('Task 1a : ', x)
    l2.acquire()
    x = x + 5
    print('Task 1b : ', x)
    l2.release()
    l1.release()

def task2(l):
    l1 = l
    l2 = l
    global x
    l2.acquire()
    print('Task 2a : ', x)
    l1.acquire()
    x = x+ 3
    print('Task 2b : ', x)
    l1.release()
    l2.release()

if __name__ == "__main__":  # confirms that the code is under main function
    lock = mp.Lock()
    procs = []
    proc1 = mp.Process(task1(lock))
    procs.append(proc1)
    proc1.start()
    proc2 = mp.Process(task2(lock))
    procs.append(proc2)
    proc2.start()

    # complete the processes
    for proc in procs:
        proc.join()

    print('Final : ', x)

```
