```r
hmeq <- read_csv('/content/hmeq.csv')
head(hmeq)
```

```r
hmeq_rev <- hmeq%>% select(-JOB)
head(hmeq_rev)
```

```r
set.seed(2020)
inTrain <- createDataPartition(y = hmeq_rev$BAD, p = 0.70, list=FALSE)
```

```r
training <- hmeq_rev[as.vector(inTrain),] 
test <- hmeq_rev[-as.vector(inTrain),]
```

```r
nrow(training)
```

```r
table(training$BAD)/nrow(training)
```

```r
nrow(test)
table(test$BAD)/nrow(test)
```

```r
hmeq01<-hmeq%>%mutate(y01=as.numeric(BAD==1)) #y01=1 if y=="yes"; otherwise y01=0.
head(hmeq01)
```

```r
hmeq_LGR<-glm(y01~DEBTINC+DELINQ+CLAGE+DEROG+VALUE, data=hmeq01, family=binomial(link = logit))
hmeq_LGR
```
