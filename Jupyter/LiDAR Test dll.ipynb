```python
import os
os.add_dll_directory(r'C:\OSGeo4W\bin')

from liblas import file
import numpy as np
import time

if __name__ == '__main__':
    start = time.perf_counter()
    f = file.File('line009.las', mode='r')
    lasPointCount = int(f.__len__())

    # X, Y, Z, scan_direction_flag, intensity, gps_time
    pArray = np.empty(shape=(lasPointCount, 6), dtype='d')
    counter = 0

    for pnt in f:
        if counter % 100000 == 0:
            end = time.perf_counter()
            print(counter, f' : Current in {round(end - start, 6)} second(s)')

        newrow = [pnt.x, pnt.y, pnt.z, pnt.scan_direction, pnt.intensity, pnt.raw_time]
        pArray[counter] = newrow
        counter += 1

    np.save('LiDAR2.npy', pArray, False)

    end = time.perf_counter()
    print('')
    print(f'Finished in {round(end - start, 2)} second(s)')

    print(len(pArray))
```
